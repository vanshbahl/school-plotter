```python
df = df.dropna()
```

```python
latitude, longitude = df['Latitude'].median(), df['Longitude'].median()
```

```python
import folium
```

```python
df['District'].value_counts()
```

```python
map1 = folium.Map(location=[latitude, longitude], zoom_start=11)

for lat, long, label, dist in zip(df['Latitude'], df['Longitude'], df['School Name'], df['District']):
    if dist == 'North East':
        c = 'blue'
    elif dist == 'North West B':
        c = 'lightgreen'
    elif dist == 'West B':
        c = 'red'
    elif dist == 'South West B':
        c = 'pink'
    elif dist == 'East':
        c = 'blue'
    elif dist == 'North West A':
        c = 'darkgreen'
    elif dist == 'South East':
        c = 'purple'
    elif dist == 'South':
        c = 'gray'
    elif dist == 'West A':
        c = 'orange'
    elif dist == 'South West A':
        c = 'cadetblue'
    elif dist == 'North':
        c = 'beige'
    elif dist == 'Central':
        c = 'white'
    elif dist == 'New Delhi':
        c = 'black'

    folium.CircleMarker(
        [lat,long],
        radius = 6,
        popup = label,
        fill=True,
        fill_color = c,
        fill_opacity=1
    ).add_to(map1)
map1
```
